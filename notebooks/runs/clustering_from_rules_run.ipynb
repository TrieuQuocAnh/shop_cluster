```python
# Parameters
CLEANED_DATA_PATH = "data/processed/cleaned_uk_data.csv"
RULES_INPUT_PATH = "data/processed/rules_fpgrowth_filtered.csv"
TOP_K_RULES = 200
SORT_RULES_BY = "lift"
WEIGHTING = "lift"
MIN_ANTECEDENT_LEN = 2
USE_RFM = True
RFM_SCALE = True
RULE_SCALE = False
K_MIN = 2
K_MAX = 10
N_CLUSTERS = 4
RANDOM_STATE = 42
OUTPUT_CLUSTER_PATH = "data/processed/customer_clusters_from_rules.csv"
PROJECTION_METHOD = "pca"
PLOT_2D = True
```

## Set up


```python
%load_ext autoreload
%autoreload 2

import os
import sys
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Determine correct project root
cwd = os.getcwd()
if os.path.basename(cwd) == "notebooks":
    project_root = os.path.abspath("..")
else:
    project_root = cwd

src_path = os.path.join(project_root, "src")
if src_path not in sys.path:
    sys.path.append(src_path)

from cluster_library import RuleBasedCustomerClusterer
```

## Load cleaned data & rules


```python
df_clean = pd.read_csv(CLEANED_DATA_PATH, parse_dates=["InvoiceDate"])
print(df_clean.shape)
df_clean.head()
```

```python
clusterer = RuleBasedCustomerClusterer(df_clean=df_clean)
customer_item_bool = clusterer.build_customer_item_matrix(threshold=1)
print('Customer × Item:', customer_item_bool.shape)

rules_df = clusterer.load_rules(
    rules_csv_path=RULES_INPUT_PATH,
    top_k=TOP_K_RULES,
    sort_by=SORT_RULES_BY,
)
print('Rules used:', rules_df.shape)
rules_df.head()
```

## Build features (Rules → Features) + (optional) RFM


```python
X, meta = clusterer.build_final_features(
    weighting=WEIGHTING,
    use_rfm=USE_RFM,
    rfm_scale=RFM_SCALE,
    rule_scale=RULE_SCALE,
    min_antecedent_len=MIN_ANTECEDENT_LEN,
)
print('X shape:', X.shape)
meta.head()
```

## Choose K (silhouette)


```python
sil_df = clusterer.choose_k_by_silhouette(
    X,
    k_min=K_MIN,
    k_max=K_MAX,
    random_state=RANDOM_STATE,
)
sil_df
```

```python
best_k = int(sil_df.loc[0, 'k'])
k = best_k if N_CLUSTERS is None else int(N_CLUSTERS)
print('Chosen k =', k)
```

## Fit KMeans & save results


```python
labels = clusterer.fit_kmeans(X, n_clusters=k, random_state=RANDOM_STATE)
meta_out = meta.copy()
meta_out['cluster'] = labels

# Lưu
os.makedirs(os.path.dirname(OUTPUT_CLUSTER_PATH), exist_ok=True)
meta_out.to_csv(OUTPUT_CLUSTER_PATH, index=False)
print('Saved:', OUTPUT_CLUSTER_PATH)
meta_out.head()
```

## Top rules theo từng cluster (Rule-based profiling)

Phần này nhằm rút ra các **dấu hiệu hành vi mua kèm đặc trưng** của từng cụm,
dựa trên các rule-features được kích hoạt nhiều nhất trong cụm.


```python
import os
import numpy as np
import pandas as pd

def top_dominant_rules_by_cluster(
    X_rules,
    rule_names,
    clusters,
    top_n=10,
    output_dir="data/processed/top_rules_rule"
):
    """
    Xác định luật TRỘI theo từng cluster và lưu ra CSV
    """
    os.makedirs(output_dir, exist_ok=True)

    results = {}

    # Tỷ lệ kích hoạt toàn bộ data
    global_rate = X_rules.mean(axis=0) + 1e-6  # tránh chia 0

    for c in np.unique(clusters):
        idx = clusters == c
        cluster_rate = X_rules[idx].mean(axis=0)

        dominance = cluster_rate / global_rate

        df = pd.DataFrame({
            "rule": rule_names,
            "activation_rate": cluster_rate,
            "global_rate": global_rate,
            "dominance": dominance
        })

        # LỌC LUẬT TRỘI
        df_dom = df.sort_values(
            by="dominance",
            ascending=False
        ).head(top_n)

        # Lưu file
        file_path = os.path.join(output_dir, f"cluster_{c}.csv")
        df_dom.to_csv(file_path, index=False)

        results[c] = df_dom

    print(f"✅ Saved dominant rules for each cluster to: {output_dir}")
    return results
```

```python
# Chuẩn bị dữ liệu
clusters = meta_out['cluster'].values

# Vì USE_RFM = True → 3 cột cuối là RFM
n_rules = X.shape[1] - 3
X_rules = X[:, :n_rules]

# Tên luật
rule_names = rules_df['rule_str'].values

# Chạy xác định LUẬT TRỘI
dominant_rules = top_dominant_rules_by_cluster(
    X_rules=X_rules,
    rule_names=rule_names,
    clusters=clusters,
    top_n=TOP_K_RULES,
    output_dir="data/processed/top_rules_rule"
)


```

## Xác định luật trội

```python
def dominant_rules_by_cluster(
    X_rules,
    rule_names,
    clusters,
    top_n=10,
    min_activation=0.1,
    min_dominance=1.2
):
    """
    X_rules: (n_customers, n_rules)
    rule_names: list[str]
    clusters: cluster labels
    """

    results = {}

    global_rate = X_rules.mean(axis=0) + 1e-6  # tránh chia 0

    for c in np.unique(clusters):
        idx = clusters == c
        cluster_rate = X_rules[idx].mean(axis=0)

        dominance = cluster_rate / global_rate

        df = pd.DataFrame({
            "rule": rule_names,
            "activation_rate": cluster_rate,
            "global_rate": global_rate,
            "dominance": dominance
        })

        # lọc luật trội
        df_dom = df[
            (df["activation_rate"] >= min_activation) &
            (df["dominance"] >= min_dominance)
        ].sort_values("dominance", ascending=False).head(top_n)

        results[c] = df_dom.reset_index(drop=True)

    return results
```

```python
dominant_rules = dominant_rules_by_cluster(
    X_rules=X_rules,
    rule_names=rule_names,
    clusters=clusters,
    top_n=TOP_K_RULES
)

dominant_rules[0]   # luật trội cluster 0
```
